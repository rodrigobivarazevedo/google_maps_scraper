```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
df = pd.read_csv('../data/Île-de-France_cleaned.csv')

df.head()
```

```python
regions_level2 = gpd.read_file("../france_boundary/gadm36_FRA_2.shp")
```

```python
# Convert the DataFrame to a GeoDataFrame
facilities = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))

# Set the coordinate reference system (CRS) to EPSG 4326
facilities.crs = {'init': 'epsg:4326'}

# View the first five rows of the GeoDataFrame
facilities.head()
```

```python
# Create a map
ax = regions_level2.plot(figsize=(20,20), color='whitesmoke', linestyle=':', edgecolor='black')
facilities.plot(markersize=5, ax=ax)
```

```python
facilities.department.value_counts()
```

```python
# Aggregate farms by region and count the number of farms in each region
farms_by_region = facilities.groupby('department')['name'].count()
farms_by_region
```

```python
# create separate GeoDataFrames for each region with a loop and store them in a dictionary
farms_by_region_dict = {}
for region in facilities['department'].unique():
    farms_by_region_dict[region] = facilities[facilities['department'] == region]

# create a list with 13 colors
colors = ['red', 'blue', 'green', 'purple', 'orange', 'pink', 'gray', 'olive', 'cyan', 'brown', 'black', 'yellow', 'magenta']

# create a list with 20 colors
#colors = plt.cm.tab20.colors

ax = regions_level2.plot(figsize=(20,20), color='whitesmoke', linestyle=':', edgecolor='black', zorder=1)
i = 0
for region, gdf in farms_by_region_dict.items():
    gdf.plot(markersize=5, color=colors[i], ax=ax, zorder=2)
    i += 1
```

```python
regions_level1 = gpd.read_file("../france_boundary/gadm36_FRA_1.shp")

# create a map with the names of the regions
ax = regions_level1.plot(figsize=(20,20), color='whitesmoke', linestyle=':', edgecolor='black', zorder=1)
regions_level1.apply(lambda x: ax.annotate(text=x.NAME_1, xy=x.geometry.centroid.coords[0], ha='center'),axis=1)
```
